```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('data_breaches.csv')

# Convert the 'Year' column to numeric values, coercing errors to NaN, then drop NaNs
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data = data.dropna(subset=['Year'])

# Now convert the 'Year' column to integers
data['Year'] = data['Year'].astype(int)

# Filter the data to include only breaches from 2020 onwards
data_since_2020 = data[data['Year'] >= 2020]

# Calculate the number of breaches per year
breaches_per_year = data_since_2020['Year'].value_counts().sort_index()

# Creating a bar chart for the number of breaches per year
plt.figure(figsize=(10, 6))
breaches_per_year.plot(kind='bar', color='royalblue')
plt.title('Number of Data Breaches Since 2020')
plt.xlabel('Year')
plt.ylabel('Number of Data Breaches')
plt.xticks(rotation=0)  # No rotation needed for year labels
plt.tight_layout()  # Adjust the layout
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('data_breaches.csv')

# Ensure 'Year' is a numeric field; this will convert non-numeric to NaN
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')

# Filter out rows with NaN 'Year'
data = data.dropna(subset=['Year'])

# Convert 'Year' to an integer type
data['Year'] = data['Year'].astype(int)

# Filter data for years 2020-2024
data_filtered = data[(data['Year'] >= 2020) & (data['Year'] <= 2024)]

# Filter organizations of interest
organizations_of_interest = ['government, military', 'financial', 'healthcare', 'ppersonal and demographic data about residents and their properties of US', 'social media']
filtered_data = data_filtered[data_filtered['Organization type'].isin(organizations_of_interest)]

# Group data by 'Organization type' and sum up the occurrences
grouped_data = filtered_data.groupby('Organization type').size()

# Plotting
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color='skyblue')
plt.title('Number of Breaches by Organization Type (2020-2024)')
plt.xlabel('Organization Type')
plt.ylabel('Number of Breaches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
