```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('Voices _from_the_Digital_Frontier.csv')

# Clean the column names by stripping extra spaces and control characters
data.columns = data.columns.str.strip()

# Verify the cleaned column names
print(data.columns)

# Use the cleaned column name for your plot
concern_levels = data['How concerned are you about the security of your personal data online?'].value_counts().sort_index()

# Creating a bar chart to visualize concern levels
concern_levels.plot(kind='bar', color='cadetblue')
plt.title('Overall Concern About Data Security')
plt.xlabel('Concern Level')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('Voices _from_the_Digital_Frontier.csv')

# Assuming you've already cleaned the column names earlier
# Find the column for question 8 based on partial matching
sharing_purposes_column = [col for col in data.columns if "purposes would you be most willing to share your personal data" in col][0]

# Split the selected purposes into a list, explode them into separate rows, and then strip whitespace
sharing_purposes = data[sharing_purposes_column].str.split(';').explode().str.strip()

# Filter out any empty or whitespace-only responses
sharing_purposes = sharing_purposes[sharing_purposes != '']

# Count the occurrences of each purpose
sharing_purposes_counts = sharing_purposes.value_counts()

# Creating a bar chart for the purposes people are willing to share personal data for
plt.figure(figsize=(12, 8))  # Adjusted figure size for better readability
sharing_purposes_counts.plot(kind='bar', color='skyblue')
plt.title('Purposes for Sharing Personal Data with Companies or Organizations')
plt.xlabel('Purposes')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)  # Rotate labels to vertical for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to avoid clipping of tick-labels
plt.show()
```
